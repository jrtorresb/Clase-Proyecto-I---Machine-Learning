# Underfitting - Overfitting

Underfitting: el modelo no captura la lógica fundamental de los datos. Tiene poca precisión.
<br>
Overfitting: el modelo se enfoca de forma muy particular y modela el ruido. Alta precisión.

Una solución para Overfitting es dividir los datos en dos conjuntos: Entrenamiento (80%) y Prueba (20%).
<br>
Creamos el modelo en el conjunto de Entrenamiento y verificamos su precisión en el conjunto de Prueba.
<br>
Si el modelo tiene una alta precisión en los datos de Entrenamiento pero baja precisión en los datos de Prueba, entonces el modelo esta Overfitting.

# Train Test Split

## Importamos bibliotecas

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

## Generamos datos

```python
# Construimos un dataframe con enteros del 1 al 100
a = np.arange(1,101)
a
```

```python
# Contruimos un dataframe con enteros del 501 al 600
b = np.arange(501,601)
b
```

## Dividimos los datos
Full documentation: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
# La división por default es 75-25
# Ambos arreglos estan revueltos.
train_test_split(a)
```

```python
# La mayoría de las veces, tenemos entradas y objetivos, por lo que tenemos que dividir 2 matrices diferentes
# estamos simulando esta situación dividiendo 'a' y 'b'

# Se puede especificar el 'test_size' 
# Las divisiones comunes son 75-25, 80-20, 85-15, 90-10

# Siempre se debe emplear un 'random_state'
# para asegurar que cuando se dividan los datos siempre obtendrá la misma partición.

# Nota: 2 matrices se dividirán en 4
# El orden es train1, test1, train2, test2
# Es muy útil almacenarlos en 4 variables, para luego poder usarlos 
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=365)
```
