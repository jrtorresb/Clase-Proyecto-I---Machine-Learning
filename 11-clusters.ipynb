# Clusters

El análisis de clusters es una técnica estadística multivariada que agrupa las observaciones sobre la base de algunas de sus características o variables que los describen. 

## Ejemplos de clusters

Tenemos 6 observaciones de paises: Canada Australia, USA, Francia, Alemania y UK.
<br>
Podemos agruparlos por:
<ul>
    <li> Proximidad geográfica:
    <li> 1) Canada y USA </li>
    <li> 2) Alemania, Francia, UK </li>
    <li> 3) Australia </li>
    </li>
    <br>
    <li> Hemisferio de localización: 
    <li> 1) Canada, US, UK, Francia, Alemania </li>
    <li> 2) Australia> </li>
    </li>
    <ul>

El objetivo del clustering es maximizar la similitud de las observaciones dentro de un grupo y maximizar la diferencia entre los grupos de acuerdo a alguna carateristica. 

## Aplicaciones

<ul>
<li> Segmentación de Mercados </li>
<li> Segmentación de Imágenes y compresión </li>
</ul>
Permite hacer una primera exploración de los datos e identificar patrones dentro de ellos.

## Recordando conceptos

## Clasificación vs Clustering

La clasificación es un ejemplo de aprendizaje supervisado. <br>
Se utiliza siempre que tenemos datos de entrada y los resultados correctos deseados (objetivos). <br>
Entrenamos nuestros datos para encontrar los patrones en las entradas que conducen a los objetivos. <br>
Con la clasificación, esencialmente necesitamos saber la clase correcta de cada una de las observaciones en nuestros datos, para poder aplicar el algoritmo. <br>
Usamos los objetivos (valores correctos) para ajustar el modelo y obtener mejores resultados. <br>
Una regresión logística es un ejemplo de clasificación.
<br> 
El análisis de clustering es un ejemplo de aprendizaje no supervisado. <br>
Se utiliza siempre que tenemos datos de entrada pero no sabemos cuáles son los resultados correctos. <br>
El clustering consiste en agrupar puntos de datos en función de las similitudes entre ellos y las diferencias con los demás.

## Tipos de Clustering

<ul> 
    <li> Flat: Con los métodos planos no hay jerarquía, se elige el número de clústeres antes de la agrupación.
        <br>
Hacer una agrupación jerárquica es mucho más lenta y costosa desde el punto de vista informático.
        <br>
Hoy en día, se prefieren los métodos planos debido al volumen de datos que normalmente intentamos agrupar. 
    </li>
    <li>
        Jerárquico: Explora (contiene) todas las soluciones. 
        <br><br>
        Los tipos jerárquicos se pueden dividir en: <br>
        Divisivos (top-down): partimos de una situación en la que todas las observaciones están en el mismo grupo. Luego dividimos este gran grupo en 2 más pequeños, continuamos con 3, 4, 5, y así sucesivamente, hasta que cada observación sea su grupo separado.<br>
Para encontrar la mejor división, debemos explorar todas las posibilidades en cada paso. 
        <br><br>
        Agolmerativos (bottom-up): el enfoque es de abajo hacia arriba. 
Para encontrar la combinación de observaciones en un grupo, debemos explorar todas las posibilidades en cada paso. 
    </li>
        </ul>

La forma más intuitiva de medir la distancia entre los datos es utilizando la distancia euclidiana y la fórmula de Pitagoras.

El Centroide es la posición media de un grupo de datos.

## K-means Clustering

```python
<ol>
<li> Escoger el número k de clusters que deseamos </li>
<li> Especificar el número de semillas. Una semilla es el centroide inicial de los clusters. </li>
<li> Asignar cada punto a un centroide basado en una medida de proximidad </li>
<li> Ajustar los centroides hasta ya no cambien los clusters</li>
</ol>
```

Pros
<ul> 
    <li> Fácil de implementar </li>
    <li> Eficiencia computacional </li>
    <li> Popular </li>
    <li> Simpre de un resultado </li>
</ul>
Cons
<ul>
    <li> Necesitamos escoger K </li>
    <li> Sensible a la inicialización de las semillas </li>
    <li> Sensible a los valores atípicos </li>
    </ul>

# Ejemplo Clustering

## Importar bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set seaborn
sns.set()
# Importamos kmeans con sklearn
from sklearn.cluster import KMeans
```

## Datos

```python
data = pd.read_csv('Country clusters.csv')
data
```

## Graficamos los puntos

```python
# Hacemos la grafica de puntos usando latitud y longitud
plt.scatter(data['Longitude'],data['Latitude'])
# Establecemos los limites de los ejes
plt.show()
```

## Seleccionamos las características

```python
# iloc es un método utilizado para 'cortar' datos en pandas
# El término usado por los pandas es 'selección por posición'
# El primer argumento identifica los renglones que queremos mantener
# El segundo - las columnas
# Al elegir las columnas, p. Ej. a: b, mantendremos las columnas a, a + 1, a + 2, ..., b-1; 
# entonces la columna b está excluida

# Vamos a seleccionar las columnas 1 y 2 de latitud y longitud
x = data.iloc[:,1:3]
x
```

## Clustering

```python
# Creamos un objeto Kmeans que recibe como parametro el numero de clusters
kmeans = KMeans(3)
```

```python
# Clusterizamos los datos
kmeans.fit(x)
```

## Resultados del Clustering

```python
# Creamos una variable que contiene los clusters
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
# Creamos un dataframe con los datos
data_with_clusters = data.copy()
# Agregamos una columna con los clusters identificados
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
# Graficamos los resultados
# c (color) es un argumento que podría codificarse con una variable
# La variable tiene valores 0,1,2, indicando a plt.scatter, que hay tres colores (0,1,2)
# Todos los puntos del grupo 0 serán del mismo color, todos los puntos del grupo 1, otro, etc.
# cmap es el mapa de colores. 
plt.scatter(data_with_clusters['Longitude'],data_with_clusters['Latitude'],
            c=data_with_clusters['Cluster'],
           cmap='rainbow')
plt.show()
```
